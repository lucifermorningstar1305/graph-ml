```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

# Definitions

## Order of a Graph

The **order of a graph** is the **number of nodes/vertices** $|V|$ present in the graph $\mathcal{G}$

## Size of a Graph

The **size of a graph** is the **number of edges/relationship** $|E|$ present in the graph $\mathcal{G}$

## Degree of a Vertex/Node

The **degree** of a vertex/node is the **number of edges** that are **adjacent** to it.


## Neighbors of Vertex

The **neighbors** of a vertex $\mathcal{v}$ in a graph $\mathcal{G}$ is the **subset** of vertex $V'$ induced by all vertices adjancent to vertex $\mathcal{v}$

## Neighborhood Graph (aka ego-graph)

The **neighborhood graph** of a vertex $\mathcal{v}$ in a graph $\mathcal{G}$ is a **subgraph** of $\mathcal{G}$ composed of **all vertices adjacent** to vertex $\mathcal{v}$ and **all edges connecting vertices adjacent** to vertex  

```python
G = nx.Graph()
V = {"Dublin", "Paris", "Milan", "Rome"}
E = [("Milan", "Dublin"), ("Milan", "Paris"), ("Paris", "Dublin"), ("Milan", "Rome")]
```

```python
G.add_nodes_from(V)
G.add_edges_from(E)
```

```python
nx.draw(G, with_labels=True)
```

```python
print(f"V = {G.nodes}")
print(f"E = {G.edges}")
```

```python
print(f"Graph Order: {G.number_of_nodes()}")
print(f"Graph Size: {G.number_of_edges()}")
```

```python
print(f"Degree of each node: { {v:G.degree(v) for v in G.nodes} }")
```

```python
print(f"Neighbors of each node: { {v:list(G.neighbors(v)) for v in G.nodes} }")
```

```python
ego_graph_milan = nx.ego_graph(G, "Milan")
print(f"Nodes: {ego_graph_milan.nodes}")
print(f"Edges: {ego_graph_milan.edges}")
nx.draw(ego_graph_milan, with_labels=True)
```

```python
ego_graph_rome = nx.ego_graph(G, "Rome")
print(f"Nodes: {ego_graph_rome.nodes}")
print(f"Edges: {ego_graph_rome.edges}")
nx.draw(ego_graph_rome, with_labels=True)
```

```python
ego_graph_paris = nx.ego_graph(G, "Paris")
print(f"Nodes: {ego_graph_paris.nodes}")
print(f"Edges: {ego_graph_paris.edges}")
nx.draw(ego_graph_paris, with_labels=True)
```

# To add new nodes and edges to a Graph

```python
new_nodes = {"London", "Madrid"}
new_edges = {("London", "Rome"), ("Madrid", "Paris")}
G.add_nodes_from(new_nodes)
G.add_edges_from(new_edges)
print(f"V = {G.nodes}")
print(f"E = {G.edges}")
nx.draw(G, with_labels=True)
```

# To remove nodes

```python
nodes_remove = {"London", "Madrid"}
G.remove_nodes_from(nodes_remove)
print(f"V={G.nodes}")
print(f"E={G.edges}")
nx.draw(G, with_labels=True)
```

# To remove edges

```python
edges_remove = [("Milan", "Dublin")]
G.remove_edges_from(edges_remove)
print(f"V={G.nodes}")
print(f"E={G.edges}")
nx.draw(G, with_labels=True)
```
